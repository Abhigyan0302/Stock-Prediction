```python
import os
raw_path = ROOT / 'data/stocknet/raw'

if (raw_path / 'stocknet-dataset').exists():
    print("StockNet already downloaded — skipping.")
else:
    print("Cloning StockNet (~1.1GB)...")
    os.system(
        f"git clone https://github.com/yumoxu/stocknet-dataset.git "
        f"{raw_path}/stocknet-dataset"
    )
    print("Done.")

# Correct paths
price_dir = raw_path / 'stocknet-dataset/price/raw'
tweet_dir = raw_path / 'stocknet-dataset/tweet/raw'
print(f"Price tickers: {len(list(price_dir.iterdir()))}")
print(f"Tweet tickers: {len(list(tweet_dir.iterdir()))}")
```

```python
import os

clone_path = ROOT / 'data/stocknet/raw/stocknet-dataset'
for root, dirs, files in os.walk(clone_path):
    # Only show 2 levels deep
    level = root.replace(str(clone_path), '').count(os.sep)
    if level < 2:
        indent = ' ' * 2 * level
        print(f"{indent}{os.path.basename(root)}/")
```

```python
import yfinance as yf
from pathlib import Path
from tqdm import tqdm

TICKERS = [
    # Tech
    "AAPL","MSFT","NVDA","GOOGL","META",
    # Healthcare
    "JNJ","PFE","ABBV","MRK","UNH",
    # Financials
    "JPM","BAC","GS","WFC","MS",
    # Energy
    "XOM","CVX","COP","SLB","EOG",
    # Consumer
    "PG","KO","PEP","WMT","COST",
    # Mixed
    "TSLA","AMZN","DIS","NFLX","AMD"
]

OUT = ROOT / 'data/supplementary/prices'

for t in tqdm(TICKERS):
    out_file = OUT / f"{t}.csv"
    if out_file.exists():
        continue
    df = yf.download(t, start="2021-01-01",
                     end="2024-12-31", auto_adjust=True, progress=False)
    df.to_csv(out_file)
    print(f"Price CSVs saved: {len(list(OUT.iterdir()))}/30")
```

```python
from huggingface_hub import hf_hub_download
from pathlib import Path

LLM_DIR  = ROOT / 'models/llm'
LLM_FILE = LLM_DIR / 'Phi-3-mini-4k-instruct-Q4_K_M.gguf'

if LLM_FILE.exists():
    print(f"Already downloaded: {LLM_FILE.stat().st_size/1e9:.2f} GB")
else:
    print("Downloading Phi-3-mini Q4_K_M (~2.4GB)...")
    hf_hub_download(
        repo_id='bartowski/Phi-3-mini-4k-instruct-GGUF',
        filename='Phi-3-mini-4k-instruct-Q4_K_M.gguf',
        local_dir=str(LLM_DIR)
    )
    print("Done.")
```
